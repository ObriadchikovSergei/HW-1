### Пример
**Условие.** Напишите программу, которая просит пользователя ввести электронную почту и проверяет корректность её введения. По условию, почтовый адрес является корректным, если в нём присутствуют один и только один символ `@` (при этом слева от этого символа должен находиться хотя бы ещё один [исключая `@`]). _Это не полный список критериев, но для учебного примера достаточно этих ограничений._ При нарушении хотя бы одного условия программа должна ответить, что такая почта не принимается, и разъяснить причину. Если почта корректна, об этом также надо сообщить пользователю.

**Алгоритм решения:**
1. Получить на вход почту в виде строки (функция `input`).
2. Проверить разные условия (условная конструкция и подходящие операторы):
 - проверить наличие ровно одного символа `@` (строковый метод `.count()`) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`);
 - проверить наличие символов перед `@` (строковый метод `.startswith()` — если начинается с `@`, то условие не выполнено) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`).
3. Сообщить пользователю, что почта корректна, если это так (функция `print()`).

**Решение:**

````python
email = input()  # шаг 1

if email.count('@') != 1:  # шаг 2-1
    print('Введённый почтовый адрес некорректен. Он должен содержать ровно одну @')
elif email.startswith('@'):  # шаг 2-2
    print('Введённый почтовый адрес некорректен. Он должен иметь хотя бы один символ перед @')
else:  # шаг 3
    print('Поздравляем, ваша почта принята!')
````

### Задача 1 (5 баллов)

Пользователь подаёт на вход программе действительные числа через пробел. Он привык пользоваться Excel, поэтому разделяет числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять» — в этом случае слово надо игнорировать). **Ваша задача** — вывести минимум и максимум последовательности через запятую с пробелом. Гарантируется, что будет введено хотя бы одно число.

**Пример:**

```
input:
-2.56 25,002 восемь

output:
-2.56, 25.002
```

**Критерии:**

- задача решена корректно (учтено оба условия — разделение чисел запятыми и ввод чисел словами), описание присутствует и соответствует решению — 5 баллов;
- задача решена корректно (учтено оба условия — разделение чисел запятыми и ввод чисел словами), описание отсутствует / задача решена удовлетворительно (учтено одно из условий), описание присутствует и соответствует решению — 4 балла;
- задача решена с серьёзными ошибками, но описание присутствует и соответствует решению — 3 балла;
- иное — 0 баллов.

*Алгоритм*
1. Принять на вход данные (input()), записать данные в переменную string;
2. Отследить все числа с запятой и заменить их на точку;
3. Разбить строку на подстроки методом .split();
4. Для чисел создать новый пустой список list_of_num, перевести строки в тип данных float (если это невозможно, то данная строка не используется: проверить, сожержат ли новые строки только числа или нет, можно с помощью метода .isdigit() ). Все числа записать в новый список (с помощью метода списка .append() );
5. Найти минимум и максимум (min и max соответственно);

6. Вывести результат (print() ).

````python
# Задача 1. Решение

string = input('Введите действительные числа через пробел: ') # Шаг 1
string = string.replace(',', '.') # Шаг 2
list_of_str = string.split() # Шаг 3

list_of_num = list() # Шаг 4

for i in list_of_str:  
  if i.replace('.', '').replace('-', '').isdigit(): 
    list_of_num.append(float(i)) 

min = list_of_num[0] # Шаг 5
max = list_of_num[0] 
for i in list_of_num: 
    if i < min:
      min = i
    if i > max:
      max = i

print(f'{min}, {max}') # Шаг 6
````
